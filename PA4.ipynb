#### 1B. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female



```python
#reload the data set
data = pd.read_excel('board2.xlsx')
```

```python
#create a new column/parameter 'Average'
data['Average'] = data[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
```

```python
#create a data frame named Mindy with the following parameters:

Mindy = data.loc[(data['Gender'] == 'Female')                  #Female as a constant (Only females are selected)
               & (data['Hometown'] == 'Mindanao')              #Mindanao as a constant (Only people from Mindanao are selected)
               & (data['Average']>=55),                        #the average grade must be greater or equal to 55
                ['Name', 'Track', 'Electronics', 'Average']]   #the included columns (name, track, electronics, average)

#output the worthy students
Mindy
```
